### torch.chunk

```python
import torch
```

```python
t_2 = torch.arange(24).reshape(4,6)
t_2.shape
```

```python
t_2
```

```python
t1 = torch.chunk(input=t_2,chunks=2,dim=0)
t1
```

```python
t1[0].shape
```

```python
t2 = torch.chunk(t_2, 2, 1)
t2
```

```python
t2[0].shape
```
